```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf
```

```python
ds=pd.read_csv('Churn_Modelling.csv')
ds
```

```python
X=ds.iloc[:,3:13].values
Y=ds.iloc[:,13].values.reshape(-1,1)
np.shape(Y)
```

```python
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
X_1=LabelEncoder()
X[:,1]=X_1.fit_transform(X[:,1])
X_2=LabelEncoder()
X[:,2]=X_2.fit_transform(X[:,2])
ohe=OneHotEncoder(categorical_features=[1])
X=ohe.fit_transform(X).toarray()
X=X[:,1:]
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X=sc.fit_transform(X)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
ann=tf.keras.models.Sequential()
ann.add(tf.keras.layers.Dense(units=6,activation='relu')) #1 hidden layer
ann.add(tf.keras.layers.Dense(units=6,activation='relu')) #2 hidden layer
ann.add(tf.keras.layers.Dense(units=1,activation='sigmoid')) #output layer
```

```python
ann.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
ann.fit(X_train,Y_train,batch_size=50,epochs=100)
```

```python
pred=ann.predict(X_test)
for i in range(len(pred)):
    if pred[i]>0.5:
        pred[i]=1
    else:
        pred[i]=0      
accuracy=0
for j in range(len(pred)):
    if pred[j]==Y_test[j]:
        accuracy+=1
print(accuracy)        
accuracy=accuracy/len(pred)
print("Accuracy = ",accuracy)
```

```python
pred=(pred>0.5)
from sklearn.metrics import confusion_matrix
c=confusion_matrix(Y_test,pred)
print(c)
```
